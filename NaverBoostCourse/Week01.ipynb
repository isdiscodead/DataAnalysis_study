### 2.2 판다스 치트 시트를 활용한 기초 익히기
panel data system의 약자 -> 대용량의 행과 열로 된 데이터를 위함 

- data frame : 2차원 행렬
- series : 1차원 데이터

```python

import pandas as pd

# data frame 표 형태 ( 2차원 )
# pd.DataFrame?

df = pd.DataFrame(
    {"a" : [4, 5, 6],
     "b" : [7, 8, 9],
     "c" : [10, 11, 12]},
    index = [1,2,3]
)

df


# series 1차원 
df["a"]
df[["a"]] # data frame
df[["a", "b"]] # data frame


# subset
df[df["a"] > 4] # 2, 3번째 값만 가져옴


# summarize data
df["a"].value_counts() # 빈도
len(df) # 길이


# reshaping
df.sort_values("a", ascending=False)
df = df.drop(["c"], axis=1) # 컬럼(열) 드랍 ... 다시 담아줘야 함


# group data
df.groupby(["a"])["b"].mean().agg(["mean", "sum", "count"])
df.groupby(["a"])["b"].describe()
pd.pivot_table(df, index="a", values="b", aggfunc="sum") # 기본적으로 평균 값 계산


# plotting
# df.plot.area()
# df.plot.bar()
```
