```python

import pandas as pd

# data frame í‘œ í˜•íƒœ ( 2ì°¨ì› )
# pd.DataFrame?

df = pd.DataFrame(
    {"a" : [4, 5, 6],
     "b" : [7, 8, 9],
     "c" : [10, 11, 12]},
    index = [1,2,3]
)

df


# series 1ì°¨ì› 
df["a"]
df[["a"]] # data frame
df[["a", "b"]] # data frame


# subset
df[df["a"] > 4] # 2, 3ë²ˆì§¸ ê°’ë§Œ ê°€ì ¸ì˜´


# summarize data
df["a"].value_counts() # ë¹ˆë„
len(df) # ê¸¸ì´


# reshaping
df.sort_values("a", ascending=False)
df = df.drop(["c"], axis=1) # ì»¬ëŸ¼(ì—´) ë“œë ... ë‹¤ì‹œ ë‹´ì•„ì¤˜ì•¼ í•¨


# group data
df.groupby(["a"])["b"].mean().agg(["mean", "sum", "count"])
df.groupby(["a"])["b"].describe()
pd.pivot_table(df, index="a", values="b", aggfunc="sum") # ê¸°ë³¸ì ìœ¼ë¡œ í‰ê·  ê°’ ê³„ì‚°


# plotting
# df.plot.area()
# df.plot.bar()
```
